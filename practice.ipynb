```python
#기존코드
import cv2
import mediapipe as mp
import os  # os 모듈 추가

def find_landmarks(img_bgr, face_mesh):
    """Normalized image space to pixel space"""
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(img_rgb)
    landmarks = []
    if results.multi_face_landmarks:
        for person in results.multi_face_landmarks:
            face = []
            for id, lm in enumerate(person.landmark):
                ih, iw, _ = img_bgr.shape
                x, y = int(lm.x * iw), int(lm.y * ih)
                face.append([x, y])
            landmarks.append(face)
    return landmarks

# 비디오 파일 경로 설정
cap = cv2.VideoCapture("jmhong_face.mp4")
# cap = cv2.VideoCapture(0)  # 웹캠을 사용할 경우 주석 해제

# 'output' 폴더 생성
output_dir = "output"
os.makedirs(output_dir, exist_ok=True)  # 폴더가 없으면 생성

with mp.solutions.face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5,
) as face_mesh:

    frame_count = 0  # 프레임 카운트를 위한 변수
    while cap.isOpened():
        success, image = cap.read()
        if not success:            
            break

        landmarks = find_landmarks(image, face_mesh)

        if not landmarks:
            print("No person detected")
            continue

        # 특정 랜드마크만 그리기
        cv2.circle(image, landmarks[0][159], 2, (255, 255, 255), cv2.FILLED)
        cv2.circle(image, landmarks[0][145], 2, (255, 255, 255), cv2.FILLED)

        # 이미지 저장 (프레임 번호를 사용하여 파일 이름 생성)
        output_path = os.path.join(output_dir, f"output_frame_{frame_count}.png")
        cv2.imwrite(output_path, image)
        print(f"Saved: {output_path}")  # 저장된 파일 경로 출력

        frame_count += 1  # 프레임 카운트 증가

cap.release()
print("Video processing completed.")
```

```python
import cv2
import mediapipe as mp
import os

def find_landmarks(img_bgr, face_mesh):
    """Normalized image space to pixel space"""
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    results = face_mesh.process(img_rgb)
    landmarks = []
    if results.multi_face_landmarks:
        for person in results.multi_face_landmarks:
            face = []
            for id, lm in enumerate(person.landmark):
                ih, iw, _ = img_bgr.shape
                x, y = int(lm.x * iw), int(lm.y * ih)
                face.append([x, y])
            landmarks.append(face)
    return landmarks

cap = cv2.VideoCapture("ld.mp4")  # 비디오 파일 경로 설정

# 'output' 폴더 생성 또는 새로운 폴더 생성
base_dir = "output"
output_dir = base_dir
count = 1

# 이미 output 폴더가 존재하면 숫자를 붙여 새로운 폴더 생성
while os.path.exists(output_dir):
    output_dir = f"{base_dir}{count}"
    count += 1

os.makedirs(output_dir)  # 최종적으로 생성된 폴더 생성

with mp.solutions.face_mesh.FaceMesh(
    max_num_faces=1,
    refine_landmarks=True,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5,
) as face_mesh:

    frame_count = 0  # 프레임 카운트를 위한 변수
    while cap.isOpened():
        success, image = cap.read()
        if not success:            
            break

        landmarks = find_landmarks(image, face_mesh)

        if not landmarks:
            print("No person detected")
            continue

        # 특정 랜드마크만 그리기
        cv2.circle(image, landmarks[0][159], 2, (255, 255, 255), cv2.FILLED)
        cv2.circle(image, landmarks[0][145], 2, (255, 255, 255), cv2.FILLED)

        # 이미지 저장 (프레임 번호를 사용하여 파일 이름 생성)
        output_path = os.path.join(output_dir, f"output_frame_{frame_count}.png")
        cv2.imwrite(output_path, image)
        print(f"Saved: {output_path}")  # 저장된 파일 경로 출력

        frame_count += 1  # 프레임 카운트 증가

cap.release()
print("Video processing completed.")
```
